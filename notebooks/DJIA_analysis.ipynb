```python
import pandas as pd
import numpy as np
import yfinance as yf
import plotly.graph_objects as go
import plotly.express as px
```

```python
DJI_df = yf.download('DJI', 
                      start='1950-01-01',
                      progress=False)
```

```python
DJI_df
```

```python
def plotly_stock_prices(dataframe, name):
    print(f'plotting between dates \n{dataframe.index.values.min()}\n{dataframe.index.values.max()}')
    fig = px.line(dataframe['Adj Close'])
    fig.update_layout(title = {'text' : f'{name} prices by period',
                                'x':0.5,
                                'xanchor' : 'center'},
                        xaxis_title = 'Period slicer',
                        yaxis_title = 'Price',
                        template = 'plotly_dark',
                        xaxis_rangeslider_visible=True,
                        width=950, height=500)

    fig.show()
```

```python
plotly_stock_prices(DJI_df, 'DJI')
```

```python
DJI_wk = DJI_df.copy()
```

```python
DJI_wk['Adj Close']
```

```python
DJI_wk['log_return'] = (np.log(DJI_wk['Adj Close']) - np.log(DJI_wk['Adj Close'].shift(30)))*100
```

```python
DJI_wk.tail()
```

```python
fig = px.histogram(DJI_wk['log_return'])
fig.update_layout(title = 'Distribution of log_retutn values',
                template = 'plotly_dark', 
                width=1000, 
                height=500)
fig.show()
```

```python
ten_year_yield = pd.read_csv('/content/data/10-year-treasury-bond-rate-yield-chart.csv')
```

```python
ten_year_yield.head()
```

```python
ten_year_yield['date'] = pd.to_datetime(ten_year_yield['date'])
```

```python
ten_year_yield.index = ten_year_yield['date']
```

```python
ten_year_yield
```

```python
ten_year_yield.drop(['date'], axis = 1, inplace = True)
```

```python
fig = px.line(ten_year_yield[' value'])
fig.update_layout(title = {'text' : 'ten year treasury yield value',
                                'x':0.5,
                                'xanchor' : 'center'},
                        xaxis_title = 'Period',
                        yaxis_title = 'Value',
                        template = 'plotly_dark',
                        xaxis_rangeslider_visible=False,
                        width=950, height=500)
fig.show()
```
